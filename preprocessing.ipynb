```python
import cv2
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
def load_images(folder):
    images=[]
    labels=[]
    for filename in os.listdir(folder):
        img=cv2.imread(os.path.join(folder,filename))
        name=filename[:filename.index('.')]
        if img is not None:
            images.append(img)
            labels.append(name)
    return images,labels
```

```python
def remove_color(images):
    gray_images=[]
    for image in images:
        hsv=cv2.cvtColor(image,cv2.COLOR_BGR2HSV)
        ret,grayscale=cv2.threshold(hsv[:,:,2],200,255,cv2.THRESH_BINARY)
        gray_images.append(grayscale)
    return gray_images
        
        
```

```python
def horizontal(letter,length):
    size=(130-length)//2
    front=np.ones((150,size))*255
    back=np.ones((150,size+length%2))*255
    letter=np.hstack((front,letter,back))

    return letter
```

```python
def segment_and_store(images,labels):
    count=[0 for i in range(26)]
    n=0
    for image in images:
        arr=np.unique(np.where(image==0)[1])
        start,end=arr[0],-1
        prev_end=0
        l=len(arr)
        chars=[]
        for i in range(l):
            if i==l-1:
                end=arr[i]
                
            elif arr[i+1]-arr[i]!=1:
                end=arr[i]
                
            if end!=-1:
                letter=image[:,start:end]
                letter=horizontal(letter,end-start)
                chars.append(letter)
                
                if i!=l-1:
                    start=arr[i+1]
                end=-1
        
        for i in range(len(chars)):
            char=labels[n][i]
            index= ord(char)-ord('A')
            if not os.path.exists('processed_data/'+str(char)):
                os.makedirs('processed_data/'+str(char))
            print("processed_data/"+str(char)+'/'+str(count[index])+'.png')
            cv2.imwrite("processed_data/"+str(char)+'/'+str(count[index])+'.png', chars[i])
            count[index]+=1
        n+=1
             
   
        
```

```python
images,labels=load_images('train')
bnw_images=remove_color(images)
segment_and_store(bnw_images,labels)

```
